```python
import pandas as pd
import seaborn as sns
import matplotlib as mtlb
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
# Q: 1: Data Preprocessing
df = pd.read_csv("Teen_Mental_Health_Dataset_all_column (1).csv")
df.head(10)
```

```python
#Standaridise the columns
text_cols = ["gender", "platform_usage", "social_interaction_level"]
for col in text_cols:
    if col in df.columns:
        df[col] = df[col].astype("string")
```

```python
numeric_like_cols = [
    "age",
    "daily_social_media_hours",
    "sleep_hours",
    "screen_time_before_sleep",
    "academic_performance",
    "physical_activity",
    "stress_level",
    "anxiety_level",
    "addiction_level",
    "depression_label",
]
for col in numeric_like_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
numeric_cols = df.select_dtypes(include="number").columns.tolist()
categorical_cols = df.select_dtypes(include=["object", "string", "category"]).columns.tolist()
```

```python
#Q:1
print("\nQ1: Numerical variables")
print(numeric_cols)

print("\nQ1: Categorical variables")
print(categorical_cols)

print("\nQ1: Data types")
print(df.dtypes)
```

```python
# Q:2: Descriptive Analysis
print("\nQ2a: Descriptive statistics for all numerical variables")
desc = df[numeric_cols].describe().T
desc["std"] = df[numeric_cols].std(numeric_only=True)
display(desc)
```

```python

highest_variation_var = df[numeric_cols].std(numeric_only=True).idxmax()
print(f"Variable with highest variation by standard deviation: {highest_variation_var}")
```

```python
print("\nQ2b: Mean vs median for daily_social_media_hours and sleep_hours")
compare_cols = ["daily_social_media_hours", "sleep_hours"]
summary_compare = pd.DataFrame({
    "mean": df[compare_cols].mean(),
    "median": df[compare_cols].median(),
    "difference(mean-median)": df[compare_cols].mean() - df[compare_cols].median(),
})
display(summary_compare)
```

```python
for col in compare_cols:
    mean_val = df[col].mean()
    median_val = df[col].median()
    diff = mean_val - median_val
    if pd.isna(diff):
        print(f"{col}: not enough data to interpret.")
    elif diff > 0:
        print(f"{col}: mean > median, suggesting right-skew / positive skew.")
    elif diff < 0:
        print(f"{col}: mean < median, suggesting left-skew / negative skew.")
    else:
        print(f"{col}: mean and median are equal, suggesting a more symmetric distribution.")
```

2: b: Intrepreattion
The mean and median of both daily social media hours and sleep hours are very close. This suggests the distributions are nearly symmetric with only a very slight positive skew, since the mean is a little higher than the median


```python
# Q3: Missing values and imputation

print("\nQ3a: Missing values by column")
missing = pd.DataFrame({
    "missing_count": df.isna().sum(),
    "missing_percent": (df.isna().mean() * 100).round(2),
})
missing = missing[missing["missing_count"] > 0].sort_values("missing_count", ascending=False)
display(missing)
```


Q3b: Suitable imputation for daily_social_media_hours, sleep_hours, screen_time_before_sleep:
For daily social media hours, sleep hours, and screen time before sleep, median imputation is suitable because these are numerical variables and the median is less affected by outliers. If needed, KNN or regression imputation can also be used for more accurate estimates.


Q3c: Suitable imputation for stress_level, anxiety_level, addiction_level:
For stress level, anxiety level, and addiction level, median imputation is also suitable since these are ordered severity scores. If the variables are treated as ordinal, ordinal-based or model-based imputation is better than mean imputation.

Q3d: MAR vs MNAR explanation:
: MAR: anxiety_level missingness may depend on observed variables such as sleep_hours, stress_level, or age
: MNAR: students with very high anxiety may be less likely to answer the anxiety question itself, so missingness depends on the unseen anxiety value

```python
# Q5: Correlation analysis

corr_cols = [col for col in numeric_cols if col != "depression_label"]
corr = df[corr_cols].corr()

print("\nQ5: Correlation matrix")
display(corr)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", center=0)
plt.title("Correlation Matrix of Numerical Variables")
plt.tight_layout()
plt.show()
```

```python
def corr_strength(r):
    a = abs(r)
    if a < 0.2:
        return "very weak"
    elif a < 0.4:
        return "weak"
    elif a < 0.6:
        return "moderate"
    elif a < 0.8:
        return "strong"
    return "very strong"
```

```python
focus_pairs = [
    ("daily_social_media_hours", "sleep_hours"),
    ("daily_social_media_hours", "stress_level"),
    ("daily_social_media_hours", "anxiety_level"),
    ("daily_social_media_hours", "addiction_level"),
    ("sleep_hours", "stress_level"),
    ("sleep_hours", "anxiety_level"),
    ("stress_level", "anxiety_level"),
    ("stress_level", "addiction_level"),
    ("anxiety_level", "addiction_level"),
    ("academic_performance", "stress_level"),
    ("academic_performance", "anxiety_level"),
]
```

```python
print("\nQ5: Targeted correlation interpretation")
for a, b in focus_pairs:
    if a in corr.columns and b in corr.columns:
        r = corr.loc[a, b]
        direction = "positive" if r > 0 else "negative"
        print(f"{a} vs {b}: r = {r:.2f} -> {corr_strength(r)} {direction} correlation")
```


Q5: Why categorical variables should not be directly included:
Categorical variables should not be directly included in correlation analysis because correlation works only with meaningful numeric values. If we assign simple numbers to categories, those numbers create a fake order and distance that do not exist, which can give misleading results. So categorical data must be encoded properly before any correlation study.


```python
# Q: 6: Outliers Detection using IQR

outlier_cols = ["daily_social_media_hours", "sleep_hours", "stress_level", "anxiety_level"]
outlier_results = []
```

```python
for col in outlier_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    outlier_results.append({
        "variable": col,
        "q1": q1,
        "q3": q3,
        "iqr": iqr,
        "lower_bound": lower,
        "upper_bound": upper,
        "outlier_count": len(outliers),
    })
```

```python

outlier_table = pd.DataFrame(outlier_results)
print("\nQ6a: IQR based outlier summary")
display(outlier_table)
```

```python
for col in outlier_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"\n{col}: {len(outliers)} outliers")
    if len(outliers) > 0:
        display(outliers[[col]].head(10))
```

Q6b: Interpretation
The extreme social media usage values do not appear to be data errors because the IQR method found no outliers. So these values are likely meaningful behavioural patterns rather than invalid entries. However, if any very high value appears along with inconsistent age or other variables, it should still be checked manually.
